### Ch1_03

```python
# A number is even if division by 2 gives a remainder of 0.
# If the remainder is 1, it is an odd number.
```

```python
num = float(input("Enter a number: "))
if num.is_integer():
    if (num % 2) == 0:
        print("{0} is even ".format(int(num)))
    else:
        print("{0} is odd ".format(int(num)))
else:
    print("{0} is not an integer ".format(int(num)))
```

### Ch1_06

```python
import numpy as np
# numpy.linspace()函数产生1～10，10个自然数
a_i = np.linspace(1,10,10)
print(a_i)
# numpy.cumsum()函数进行累计求和
a_i_cumsum = np.cumsum(a_i)
print(a_i_cumsum)
```

## 1.4 向量：数字排成行、列

**行向量**：排成表格的一行数字。  
**列向量**：排成表格的一列数字。

"1x3"表示1行3列，"3x1"表示3行1列  
  
**利用Numpy库**：

```python
import numpy
print(numpy.array([[1,2,3]]))
print(numpy.array([[1],[2],[3]]))
```

**转置**：行向量转置得列向量，列向量转置得行向量  


### Ch1_08

```python
import numpy as np
# row vector transposed to a column vector
a_row = np.array([[1,2,3]])
b = a_row.T
b_col = np.array([[1],[2],[3]])
print(b)
print(b_col)
```
